# Day 02: Linear Regression vs Logistic Regression

## Objective
To understand the difference between Linear Regression and Logistic Regression using a simple example.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LogisticRegression
```

```python
# Input data
X = np.array([[1], [2], [3], [4], [5]])

# Linear Regression output (numbers)
y_linear = np.array([2, 4, 6, 8, 10])

# Logistic Regression output (0 or 1)
y_logistic = np.array([0, 0, 0, 1, 1])
```

```python
linear_model = LinearRegression()
linear_model.fit(X, y_linear)

prediction = linear_model.predict([[6]])
print("Linear Regression Prediction for 6:", prediction)
```

```python
logistic_model = LogisticRegression()
logistic_model.fit(X, y_logistic)

prediction = logistic_model.predict([[6]])
print("Logistic Regression Prediction for 6:", prediction)
```

## Explanation
- Linear Regression is used to predict continuous values like price or marks.
- Logistic Regression is used to predict categories like yes/no or spam/not spam.


```python
plt.scatter(X, y_linear, label="Linear Data")
plt.plot(X, linear_model.predict(X), color="red")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression")
plt.show()
```

## Conclusion
Linear Regression predicts numerical values, while Logistic Regression predicts class labels.
