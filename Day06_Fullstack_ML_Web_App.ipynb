# Day 06: Full-Stack ML Integrated Web Application

## Objective
To demonstrate how a Machine Learning model can be integrated with a web application using Flask.


```python
!pip install flask
```

```python
# Simple ML logic (no training needed)

def predict_sentiment(text):
    if "good" in text.lower() or "love" in text.lower():
        return "Positive"
    else:
        return "Negative"
```

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def home():
    return "ML Web App is Running"

@app.route("/predict/<text>")
def predict(text):
    result = predict_sentiment(text)
    return f"Sentiment: {result}"

app
```

## Explanation
- A simple sentiment prediction logic was created.
- Flask was used to build a web application.
- The ML logic was integrated into the web route.

## Conclusion
This demonstrates how Machine Learning models can be integrated into web applications using Flask.
